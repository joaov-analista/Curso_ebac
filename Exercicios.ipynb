```python
# receber o numero

print('==================')
print('Operações')
print('1 - Adição')
print('2 - Subtração')
print('3 - Divisão')
print('4 - Multiplicação')
print('5 - Porcentagem')
print('==================')
operacao = int(input('Digite a operação que deseja fazer: '))
# qtd de numeros somados
qtd = int(input('Digite a quantidade de números que deseja calcular: '))

list = []
for n in range(qtd):
  var = float(input('Digite o número: '))
  list.append(var)

# condiçoes para as operações
# soma
if operacao == 1:
  soma = sum(list)
  print('A soma é de: ', soma)
# subtracao
elif operacao == 2:
  sub = list[0]
  for n in list[1:]:
    sub -= n
  print('A subtração é de: ', sub)
# divisao
elif operacao == 3:
  div = list[0]
  for n in list[1:]:
    if n == 0:
      print('Não é possivel dividir por 0')
      break
    else:
      div /= n
  print('A divisão é de: ', div)
# multiplicacao
elif operacao == 4:
  mult = 1
  for n in list:
    mult *= n
  print('A multiplicação é de: ', mult)
# porcentagem
elif operacao == 5:
  percent = [0]
  for n in list[1:]:
    percent = (percent * n) / 100
  print('A porcentagem é de: ', percent)
# opcao invalida na operacao
#  else:
 #   print('Operação invalida')
```

```python
# menu
print('==================')
print('Operações')
print('1 - Adição')
print('2 - Subtração')
print('3 - Divisão')
print('4 - Multiplicação')
print('5 - Porcentagem')
print('6 - Média')
print('==================')
operacao = int(input('Digite a operação que deseja fazer: '))
# opcao invalida
if operacao > 6:
  print('Opção invalida')
# valores que serão calculados
else:
  num1 = float(input('Digite o primeiro número: '))
  num2 = float(input('Digite o segundo número: '))

# soma
if operacao == 1:
  soma = num1 + num2
  print('A soma é de: ', soma)
# subtracao
elif operacao == 2:
  sub = num1 - num2
  print('A subtração é de: ', sub)
# divisao
elif operacao == 3:
  div = num1 / num2
  print('A divisão é de: ', div)
# multiplicacao
elif operacao == 4:
  multi = num1 * num2
  print('A multiplicação é de: ', multi)
# porcentagem
elif operacao == 5:
  percent = (num1 / num2) * 100
  print('A porcentagem é de: ', percent)
elif operacao == 6:
  media = (num1 + num2) / 2
  print('A média é de: ', media)
```

```python
def calculadora():
  print('==================')
  print('Operações')
  print('1 - Adição')
  print('2 - Subtração')
  print('3 - Divisão')
  print('4 - Multiplicação')
  print('5 - Porcentagem')
  print('6 - Média')
  print('0 - Sair')
  print('==================')

  while True:
    escolha = input('Escolha uma operação: ')
    if escolha == '0':
      print("Encerrando...")
      break
    if escolha in ('0', '1', '2', '3', '4', '5', '6'):
      num1 = float(input('Digite o primeiro número: '))
      num2 = float(input('Digite o segundo número: '))
      if escolha == '1':
        soma = num1 + num2
        print('A soma é de: ', soma)
      elif escolha == '2':
        sub = num1 - num2
        print('A subtração é de: ', sub)
      elif escolha == '3':
        if num2 == 0:
          print('Não é possivel dividir por 0')
        else:
          div = num1 / num2
        print('A divisão é de: ', div)
      elif escolha == '4':
        multi = num1 * num2
        print('A multiplicação é de: ', multi)
      elif escolha == '5':
        percent = (num1 / num2) * 100
        print('A porcentagem é de: ', percent)
      elif escolha == '6':
        media = (num1 + num2) / 2
        print('A média é de: ', media)
  else:
      print('Opção invalida')

calculadora()
```
